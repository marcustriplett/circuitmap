## Lasso map

```python
def find_common(arrs):
    """Recursive algorithm for finding common elements in a list of arrays.
    """
    if len(arrs) == 1:
        return arrs[0]
    elif len(arrs) == 2:
        return np.intersect1d(arrs[0], arrs[1])
    else:
        return np.intersect1d(arrs[0], find_common(arrs[1:]))

def find_reference(loc, reference, secondary=None):
    common = find_common([np.where(loc[i] == reference[:, i])[0] for i in range(len(loc))])
    if secondary is None:
        return common
    else:
        return np.intersect1d(common, secondary)
    
def find_stim_id(loc):
    return find_reference(loc, targets)

def lasso_map(power):
    trial_locs = np.where(I_multicell == power)[0]
    cols = np.array([find_stim_id(L_multicell[trial_locs[i]]) for i in range(len(trial_locs))]).squeeze()
    rows = np.array(range(len(cols)))
    sparse_data = np.ones(len(cols))
    A = csc_matrix((sparse_data, (rows, cols)))
    
    alpha = 1e-4
    sparse_lasso = Lasso(alpha=alpha, fit_intercept=False, max_iter=1000)
    sparse_lasso.fit(A, y_multicell[trial_locs])
    
    return sparse_lasso.coef_
    
```

```python
lmap50 = lasso_map(50)
lmap70 = lasso_map(70)
```

```python
fname = '210422_cell5_LP_D_A'
data = np.load('%s.npz'%fname)

y, I, L = data['y'], data['I'], data['L']

xr, yr, zr = np.unique(L[:, 0]), np.unique(L[:, 1]), np.unique(L[:, 2])
dims = [len(arr) for arr in [xr, yr, zr]]

xgrid, ygrid, zgrid = np.meshgrid(xr, yr, zr)
grid = np.c_[xgrid.flatten(), ygrid.flatten(), zgrid.flatten()]
```

```python
depths = np.unique(L[:, -1]); ndepths = len(depths)
powers = np.unique(I); npowers = len(powers)
```

```python
def find_common(arrs):
    """ Recursively finds common elements between list of arrays.
    """
    return arrs if len(arrs) <= 1 else np.intersect1d(arrs[0], find_common(arrs[1:]))

def get_receptive_field(y, I, L):

    xs, ys, zs = [np.unique(L[:, i]) for i in range(L.shape[1])]
    xdim, ydim, zdim = len(xs), len(ys), len(zs)
    powers = np.unique(I)
    npowers = len(powers)
    receptive_field = np.zeros((xdim, ydim, zdim, npowers))

    for ix, xv in enumerate(xs):
        for iy, yv in enumerate(ys):
            for iz, zv in enumerate(zs):
                for ip, pv in enumerate(powers):
                    vals = [xv, yv, zv]
                    xlocs, ylocs, zlocs = [np.where(L[:, i] == vals[i])[0] for i in range(len(vals))]
                    plocs = np.where(I == pv)[0]
                    common_indices = find_common([xlocs, ylocs, zlocs, plocs])
                    receptive_field[ix, iy, iz, ip] = np.mean(y[common_indices])
    return receptive_field
```

```python
z0 = 24
omega0 = 8
I_th = 20


def S_0(r, z, z0, omega0):
    return jnp.exp(-2*r**2/omega0**2) * jnp.exp(-2*z**2/z0**2)

def S_S(r, z, I0, I_th, z0, omega0):
    alpha = jax.nn.relu(I0**2 - I_th**2)/I0**2
    zS = jnp.sqrt(alpha * z0**2 + 1e-8)
    omegaS = jnp.sqrt(alpha * omega0**2 + 1e-8)
    return jnp.exp(-2*r**2/omegaS**2) * jnp.exp(-2*z**2 / zS**2)

def Se(r, z, I, I_th=I_th, z0=z0, omega0=omega0):
    Is = jnp.sqrt(jax.nn.relu(I**2 - I_th**2))
    return (I**2 * S_0(r, z, z0, omega0) - (I > I_th) * Is**2 * S_S(r, z, I, I_th, z0, omega0))/I_th**2

def dist_euclid(vec1, vec2):
    return jnp.sqrt(jnp.sum(jnp.square(vec1 - vec2), 1))

def dist_euclid_1d(vec1, vec2):
    return jnp.abs(vec1 - vec2)

def _psf(g, I, center, I_th=I_th, z0=z0, omega0=omega0):
    return Se(dist_euclid(g[:, :2], center[:2]), dist_euclid_1d(g[:, -1], center[-1]), I, z0=z0, omega0=omega0, I_th=I_th)

# psf = jit(vmap(partial(vmap(_psf, in_axes=(0, None, None)), grid), in_axes=(0, 0)))
```

```python
@jit
def negloglik_with_barrier(y, filt, psfc, t):
    lam = jax.nn.sigmoid(psfc @ filt)
    return -jnp.sum(y * jnp.log(lam + 1e-13) + (1 - y) * jnp.log(1 - lam + 1e-13)) - jnp.sum(jnp.log(filt))/t

@jit
def newton_method(y, psfc, niters=20, t=1e1, backtrack_alpha=0.25, backtrack_beta=0.5, max_backtrack_iters=40):
    def backtrack_cond(carry):
        it, _, lhs, rhs, _, _, _ = carry
        return jnp.logical_and(it < max_backtrack_iters, jnp.logical_or(jnp.isnan(lhs), lhs > rhs))

    def backtrack(carry):
        it, step, lhs, rhs, v, J, filt = carry
        it += 1
        step *= backtrack_beta
        lhs = negloglik_with_barrier(y, filt + step * v, psfc, t)
        rhs = negloglik_with_barrier(y, filt, psfc, t) + backtrack_alpha * step * J @ v
        return (it, step, lhs, rhs, v, J, filt)

    def get_ineq(y, filt, step, v, psfc, t, J, backtrack_alpha):
        return negloglik_with_barrier(y, filt + step * v, psfc, t), negloglik_with_barrier(y, filt, psfc, t) + backtrack_alpha * step * J @ v

    @jit
    def get_stepv(filt, args, t):
        y, psfc = args
        lam = jax.nn.sigmoid(psfc @ filt)
        J = (lam - y) @ psfc - 1/(t * filt)
        H = jnp.einsum('ij,ik->jk', (lam * (1 - lam))[:, None] * psfc, psfc) + jnp.diag(1/(t*filt**2))
        H_inv = jnp.linalg.inv(H)
        v = -H_inv @ J
        return v, J
    
    @jit
    def newton_step(filt, _):
        args = [y, psfc]
        v, J = get_stepv(filt, args, t)  
        step = 1.
        lhs, rhs = get_ineq(y, filt, step, v, psfc, t, J, backtrack_alpha)
        init_carry = (0, step, lhs, rhs, v, J, filt)
        carry = while_loop(backtrack_cond, backtrack, init_carry)
        _, step, lhs, _, _, _, _ = carry
        filt += step * v
        return filt, lhs
    
    key = jax.random.PRNGKey(1)
    filt = jax.random.uniform(key, shape=[dfilt])
    return scan(newton_step, filt, jnp.arange(niters))
```

```python
prior_variance = 1e0
filt_cov = prior_variance * np.eye(dfilt)
filt_cov[-1, -1] = 1e3
filt_cov = jnp.array(filt_cov)
```

```python
@jit
def negloglik_with_barrier(y, filt, psfc, prec, t):
    lam = jax.nn.sigmoid(psfc @ filt)
    return -jnp.sum(y * jnp.log(lam + 1e-13) + (1 - y) * jnp.log(1 - lam + 1e-13)) + 1/2 * filt @ prec @ filt - jnp.sum(jnp.log(filt))/t

@jit
def laplace_approx(y, psfc, filt_cov, niters=20, t=1e3, backtrack_alpha=0.25, backtrack_beta=0.5, max_backtrack_iters=40):
    def backtrack_cond(carry):
        it, _, lhs, rhs, _, _, _, _ = carry
        return jnp.logical_and(it < max_backtrack_iters, jnp.logical_or(jnp.isnan(lhs), lhs > rhs))

    @jit
    def backtrack(carry):
        it, step, lhs, rhs, v, J, filt, prec = carry
        it += 1
        step *= backtrack_beta
        lhs, rhs = get_ineq(y, filt, step, v, psfc, t, J, backtrack_alpha, prec)
        return (it, step, lhs, rhs, v, J, filt, prec)

    def get_ineq(y, filt, step, v, psfc, t, J, backtrack_alpha, prec):
        return negloglik_with_barrier(y, filt + step * v, psfc, prec, t), negloglik_with_barrier(y, filt, psfc, prec, t) + backtrack_alpha * step * J @ v

    @jit
    def get_stepv(filt, args, t):
        y, psfc, prec = args
        lam = jax.nn.sigmoid(psfc @ filt)
        J = (lam - y) @ psfc + 1/2 * filt @ prec - 1/(t * filt)
        H = jnp.einsum('ij,ik->jk', (lam * (1 - lam))[:, None] * psfc, psfc) + 1/2 * prec + jnp.diag(1/(t*filt**2))
        H_inv = jnp.linalg.inv(H)
        v = -H_inv @ J
        return v, J, H, H_inv
    
    @jit
    def newton_step(filt_posterior, _):
        filt, prec = filt_posterior
        args = [y, psfc, prec]
        v, J, prec, cov = get_stepv(filt, args, t)  
        step = 1.
        lhs, rhs = get_ineq(y, filt, step, v, psfc, t, J, backtrack_alpha, prec)
        init_carry = (0, step, lhs, rhs, v, J, filt, prec)
        carry = while_loop(backtrack_cond, backtrack, init_carry)
        _, step, lhs, _, _, _, _, _ = carry
        filt += step * v
        return (filt, prec), lhs
    
    key = jax.random.PRNGKey(1)
    filt = jax.random.uniform(key, shape=[dfilt])
    prec = jnp.linalg.inv(filt_cov)
    filt_prior = (filt, prec)
    return scan(newton_step, filt_prior, jnp.arange(niters))
```
